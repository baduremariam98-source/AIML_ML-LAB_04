### Program-3 to practice creating and customizing different types of plots using Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Simple Line Plot
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 4, 9, 16, 25])
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, y, color='blue', linestyle='--', marker='o', label='y = x^2')
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2. Multiple Lines
```

```python
y2 = x**3
plt.figure(figsize=(6,4))
plt.plot(x, y, label='y = x^2')
plt.plot(x, y2, label='y = x^3', color='red')
plt.title("Multiple Lines")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
# 3. Scatter Plot
```

```python
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
```

```python
sizes = np.random.randint(20, 200, 50) 
colors = np.random.rand(50)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x_scatter, y_scatter, s=sizes, c=colors, alpha=0.6, cmap='viridis')
plt.title("Scatter Plot")
plt.colorbar(label='Color scale')
plt.show()
```

```python
# 4. Bar Plot
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]
```

```python
plt.figure(figsize=(6,4))
plt.bar(categories, values, color='orange')
plt.title("Bar Plot")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python
# 5. Horizontal Bar Plot
```

```python
plt.figure(figsize=(6,4))
plt.barh(categories, values, color='green')
plt.title("Horizontal Bar Plot")
plt.xlabel("Values")
plt.ylabel("Category")
plt.show()
```

```python
# 6. Histogram
```

```python
data = np.random.randn(1000) # Normally distributed data
plt.figure(figsize=(6,4))
plt.hist(data, bins=30, color='purple', edgecolor='black', alpha=0.7)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# 7. Pie Chart
```

```python
sizes = [25, 35, 20, 20]
labels = ['Python', 'C++', 'Java', 'Other']
colors = ['gold', 'lightblue', 'lightgreen', 'pink']
explode = (0.1, 0, 0, 0) # Highlight first slice
```

```python
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, explode=explode,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Pie Chart")
plt.show()
```

```python
# 8. Subplots
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1) # 1 row, 2 columns, plot 1
plt.plot(x, y1, 'b-', label='sin(x)')
plt.title('Sine Wave')
plt.legend()
```

```python
plt.subplot(1,2,2) # 1 row, 2 columns, plot 2
plt.plot(x, y2, 'r--', label='cos(x)')
plt.title('Cosine Wave')
plt.legend()
```

```python
plt.tight_layout()
plt.show()
```

```python
# 9. Customization
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, y1, color='magenta', linestyle='-.', linewidth=2, marker='s', markersize=5)
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True, linestyle=':', linewidth=1)
plt.show()
```

```python
# 10. Saving Plots
```

```python
plt.figure()
plt.plot(x, y1)
plt.title("Plot to be Saved")
plt.savefig("sine_wave_plot.png") 
plt.close() 
```
